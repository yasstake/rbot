## Binance

メモリにある直近の約定履歴取得
https://binance-docs.github.io/apidocs/spot/en/#recent-trades-list

DBに格納された約定履歴取得（要認証）
https://binance-docs.github.io/apidocs/spot/en/#old-trade-lookup-market_data


```python
BINANCE_REST_BASE = 'https://api.binance.com'

rest_get(BINANCE_REST_BASE + "/api/v3/trades?symbol=BTCBUSD&limit=10")
```

```python
rest_get(BINANCE_REST_BASE + "/api/v3/historicalTrades?symbol=BTCBUSD&limit=10")
```
